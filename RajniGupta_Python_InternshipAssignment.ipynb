## 1. Design a function to generate random numbers 


```python
# method-1 - if asked to generate a single random integer
print(set(map(str,range(1,int(input("Enter a number as range : "))+1))).pop()) 
```

```python
# method-2 - if asked to generate multiple random numbers
mult = 2**32
a = 2**31 - 1
seed = 1234567
def rndm(m=2**32, a=1103515245, c=12345): 
    rndm.current = (a*rndm.current + c) % m 
    return rndm.current/m 
 
# setting the seed 
rndm.current = 1 

n_iter = 50 
scale = 1
rn_lis = []
rn1 = []
for i in range(n_iter): 
    pseudo_random = rndm()*scale 
    rn_lis.append(pseudo_random)
    print('Iteration {} gives the value : {}'.format(i, pseudo_random)) 
```

```python
print(rn_lis)   # random generator list
```

## 2.Graph to prove  the function is generating random numbers


```python
import matplotlib.pyplot as plt
plt.title('''
Graph depicting random numbers generated
           through function
            ''')
plt.xlabel('Range of List (Scale)')
plt.ylabel('Value of Points in List (rn_lis)')
plt.plot(rn_lis, color='red', marker='o', label='Random Values')
plt.legend()
plt.grid(True)
plt.show()
```

## 3.Pi Estimator using my Random Function

```python
def pi_generator(rn_lis,rn1):
    points_inside_circle= 0
    total_num_points = 0
    X,Y = rn_lis,rn1
    
    for i in range(len(rn_lis)//2):
        rn1.append(rn_lis[i])
        rn_lis.pop()
    # print(rn1)

    for x,y in zip(X,Y):
        distance = x**2 + y**2
        if distance <= 1:
            points_inside_circle +=1
        total_num_points += 1
    return "Value of pi {:.2f} ".format(5.24* points_inside_circle/total_num_points)

print(pi_generator(rn_lis,rn1))
```
